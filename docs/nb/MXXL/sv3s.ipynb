```python
cat = Table.read('/global/cscratch1/sd/mjwilson/desi/BGS/lumfn/MXXL/galaxy_catalogue_sv3s.fits')
cat
```

```python
pl.plot(cat['ZCOS'], cat['Z'] / cat['ZCOS'], marker=',', lw=0.0)

pl.axhline(0.9, c='k', lw=0.5)
pl.axhline(1.1, c='k', lw=0.5)

pl.xlabel('ZCOS')
pl.ylabel('(Z / ZCOS)')

pl.ylim(0.5, 2.0)
```

```python
uns = np.unique(cat['NMOCK'].data)
uns
```

```python
colors = plt.rcParams['axes.prop_cycle'].by_key()['color']
```

```python
for un, color in zip(uns, colors):
    isin = cat['NMOCK'] == un

    pl.plot(cat['MOCKRA'][isin], cat['MOCKDEC'][isin], c=color, marker=',', lw=0.0, markersize=1)
    
pl.xlim(360., 0.0)
    
pl.xlabel('RA [deg.]')
pl.ylabel('DEC [deg.]')
```

```python
pl.plot(cat['RA'].data, cat['DEC'].data, c='k', marker=',', lw=0.0, markersize=1)

pl.xlim(300., 140.)
```
